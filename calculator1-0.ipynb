```python
import tkinter as tk

# DEFINE MATHEMATICAL OPERATION
def addition(x, y):
    return x + y

def subtraction(x, y):
    return x - y

def multiplication(x, y):
    return x * y

def division(x, y):
    if y == 0:
        return "Error"
    return x / y

def percentage(x):
    return x / 100

def reciprocal(x):
    if x == 0:
        return "Error"
    return 1 / x

def square(x):
    return x ** 2

def square_root(x):
    return x ** 0.5

def sign_change(x):
    return x * (-1)

# MANAGING NUMBER CLICKS
def button_click(value):
    current = operation_display.get()
    operation_display.set(current + str(value))

# MANAGING MATHEMATICAL SIGNS
def operator_click(op):
    global first_number
    global operator
    first_number = float(operation_display.get())  
    operator = op  
    operation_display.set("")  

# DEF OPERATIONS
def calculate():
    second_number = float(operation_display.get())
    if operator == "+":
        result = addition(first_number, second_number)
    elif operator == "-":
        result = subtraction(first_number, second_number)
    elif operator == "*":
        result = multiplication(first_number, second_number)
    elif operator == "/":
        result = division(first_number, second_number)
    
    operation_display.set(result)  

# CLEAR DISPLAY
def clear_display():
    operation_display.set("")
  
#CE
def canc():
    current_value = operation_display.get()
    if current_value:
        new_value = current_value[:-1]  
        operation_display.set(new_value)

#canc
def clear_entry():
    operation_display.set("")

# DEF SPECIAL FUNCTIONS
def special_function(func):
    number = float(operation_display.get())
    result = func(number)
    operation_display.set(result)

# ROOT LAYOUT
root = tk.Tk()
root.title("Calculator")
root.geometry("400x500")


operation_display = tk.StringVar()


display = tk.Entry(root, textvariable=operation_display, font=('Helvetica', 20), background='gray')
display.pack()


buttons_frame = tk.Frame(root)
buttons_frame.pack()


buttons = [
    ['%', 'CE', 'C', 'canc'],
    ['1/x', 'x^2', 'sq. root', '/'],
    ['7', '8', '9', '*'],
    ['4', '5', '6', '-'],
    ['1', '2', '3', '+'],
    ['+/-', '0', ',', '=']
]


special_functions = {
    '1/x': reciprocal,
    'x^2': square,
    'sq. root': square_root,
    '+/-' : sign_change,
    

}

# BUTTONS
for r, row in enumerate(buttons):
    for c, button_text in enumerate(row):
        if button_text in special_functions:
            button = tk.Button(buttons_frame, text=button_text, height=3, width=7, relief="raised", borderwidth=5,
                               command=lambda func=special_functions[button_text]: special_function(func))
        elif button_text == "=":
            button = tk.Button(buttons_frame, text=button_text, height=3, width=7, relief="raised", borderwidth=5, command=calculate)
        elif button_text in ["+", "-", "*", "/"]:
            button = tk.Button(buttons_frame, text=button_text, height=3, width=7,relief="raised", borderwidth=5,
                               command=lambda op=button_text: operator_click(op))
        elif button_text == "C":
            button = tk.Button(buttons_frame, text=button_text, height=3, width=7, relief="raised", borderwidth=5,command=clear_display)
        elif button_text == "CE":
            button = tk.Button(buttons_frame, text='CE', height=3, width=7, relief="raised", borderwidth=5, command=clear_entry)
        elif button_text == "canc":
            button = tk.Button(buttons_frame, text='canc', height=3, width=7,relief="raised", borderwidth=5, command=canc)
        else:
            button = tk.Button(buttons_frame, text=button_text, height=3, width=7,relief="raised", borderwidth=5,
                               command=lambda value=button_text: button_click(value))
        button.grid(row=r, column=c, padx=5, pady=5)

root.mainloop()
```
