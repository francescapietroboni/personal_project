```python
import numpy as np

to_do_list = []  #create an ampty list
Yourname = input("Insert your name here: ")

#let's define the action that this to-do list can perform

# 1: add a task to the list
def addtask():
    item = input("Enter the task to add: ")
    to_do_list.append(item)  
    print(f"This '{item}' has been added to your list")

#display the list
def listtask():
    if not to_do_list:
        print(f"Your list '{Yourname}' doesn't have any tasks yet")
    else:
        print(f"{Yourname}'s to-do list:")
        for index, item in enumerate(to_do_list):
            print(f"Item #{index + 1}: {item}")

#delete a task
def deletetask():
    listtask()
    try:
        deletetask_index = int(input("What task number do you want to delete? ")) - 1
        if 0 <= deletetask_index < len(to_do_list):
            removed_task = to_do_list.pop(deletetask_index)
            print(f"This '{removed_task}' has been removed")
        else:
            print("This task was not found in your list")
    except (ValueError, IndexError):
        print("Warning: The provided input is invalid")

#modify an existing task 
def modifytask():
    listtask()
    try:
        modifytask_index = int(input("What task number do you want to modify? ")) - 1
        if 0 <= modifytask_index < len(to_do_list):
            print(f"Current task: '{to_do_list[modifytask_index]}'")
            new_task = input("Write your new task: ")
            to_do_list[modifytask_index] = new_task
            print(f"This task has been modified to '{new_task}'")
        else:
            print("This task was not found in your list")
    except (ValueError, IndexError):
        print("Warning: The provided input is invalid")

#let's define the main funtion 
if __name__ == "__main__":
    print(f"Hello {Yourname}, welcome to your to-do list")
    print("~~~~~~~~~~~~")
    while True:
        print("\n")
        print(f"Please, {Yourname}, select one of these options:") 
        #let's assing a number to each action 
        print("1. Add a new task")
        print("2. View your tasks")
        print("3. Modify a current task")
        print("4. Delete a current task")
        print("5. Quit")

        choice = input("Select your choice: ")

        if choice == "1":
            addtask()
        elif choice == "2":
            listtask()
        elif choice == "3":
            modifytask()
        elif choice == "4":
            deletetask()
        elif choice == "5":
            break
        else:
            print("Invalid option. Try again")
    print(f"Goodbye, {Yourname}")
```
